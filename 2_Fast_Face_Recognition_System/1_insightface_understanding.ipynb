## 1. Congigure Insight Face Model

```python
from insightface.app import FaceAnalysis
```

```python
#buffalo_l model
```

```python
app_l = FaceAnalysis(name='buffalo_l',
                     root='insightface_model',
                     providers=['CPUExecutionProvider'])
app_l.prepare(ctx_id=0, det_size=(640,640))
```

```python
app_sc = FaceAnalysis(name='buffalo_sc',
                     root='insightface_model',
                     providers=['CPUExecutionProvider'])
app_sc.prepare(ctx_id=0, det_size=(640,640))
```

### 2. Load images

```python
img =cv2.imread('test-image-2.jpg')
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 3. Apply image to insightface model and get the result

### 4. Apply image to insightface model (sc) and get the result

```python
results_sc = app_sc.get(img)
```

```python
#draw the bounding box of each detected face
img_copy = img.copy()
for res in results_sc:
    x1, y1, x2, y2 = res['bbox'].astype(int)

    #draw rectange
    cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)

    #Key points
    kps = res['kps'].astype(int)
    for k1, k2 in kps:
        cv2.circle(img_copy,(k1, k2),2,(0,255,255),-1)
    #Detection score
    score = "score: {}%".format(int(res['det_score']*100))
    for d1, d2 in kps:
        cv2.putText(img_copy,score,(x1,y1),cv2.FONT_HERSHEY_DUPLEX, 1, (255,0,0))

    
cv2.imshow('bbox',img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
